1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=sp.symbols ('x')

f=10*x-5
df=sp.diff(f, x)

print("Похідна функції 10x-5:", df)
```

```python
x=np.linspace(0,10,10)
plt.plot(x, 10*x-5, label='функція 10x-5')
plt.plot(x, np.full_like(x, 10), label='похідна від функції 10x-5')


plt.title("Графік функції та її похідної")
plt.legend()
plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
x=sp.symbols ('x')

f_sin=sp.sin(x)
df_sin=sp.diff(f_sin, x)
print (f'Похідна від функціі sin(x): ', df_sin)
f_ln=sp.ln(x)
df_ln=sp.diff(f_ln, x)
print (f'Похідна від функціі ln(x): ', df_ln)
```

```python
x = np.linspace(-10, 10, 400)

f = np.sin(x)
df = np.cos(x)

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='sin(x)')
plt.plot(x, df, label=df_sin)


plt.title("Графік функції sin(x) та її похідної")
plt.legend()
plt.show()
```

```python
x = np.linspace(-10, 10, 400)

f = np.log(x)
df = 1/x

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='ln(x)')
plt.plot(x, df, label=df_ln)

plt.title("Графік функції ln(x) та її похідної")
plt.legend()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x=sp.symbols ('x')

f=x**2-5*x+1
df=sp.diff(f, x)
print (f'Похідна від функціі: ', df)
```

```python
x = np.linspace(0, 100, 1000)

f = x**2-5*x+1
df = 2*x-5

plt.figure(figsize=(10, 5))
plt.plot(x, f, label='x**2-5*x+1')
plt.plot(x, df, label="2*x-5")

plt.title("Графік функції та її похідної")
plt.legend()
plt.show()
```
