Виведіть додатково графік розсіювання викликавши код нижче і зробіть заключення, чи варто щось змінити при підрахунку кореляції, аби коефіцієнт Пірсона показував коректну оцінку лінійної залежності?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

```python
mean_y = sum(visits)/(len(visits)-1)
y_minus_mean = [(i - mean_y)**2 for i in visits]
std_y = (sum(y_minus_mean)/(len(visits)-1))**0.5
std_y
```

```python
np.std(visits, ddof = 1)
```

```python
mean_y_2 = sum(visits)/(len(visits))
y_minus_mean_2 = [(i - mean_y_2)**2 for i in visits]
std_y_2 = (sum(y_minus_mean_2)/(len(visits)-1))**0.5
std_y_2
```

```python
cov_prep = [ (sum_purchases[i] - mean_x)*(visits[i] - mean_y) for i in range(len(sum_purchases))]
cov = sum(cov_prep)/(len(sum_purchases)-1)

cov/(std_x*std_y)
```

```python
np.corrcoef(sum_purchases, visits)
```

```python
# Так як кореляції не співпадають, розраховую з використанням звичайного середнього (без зміщення)

cov_prep_2 = [ (sum_purchases[i] - mean_x_2)*(visits[i] - mean_y_2) for i in range(len(sum_purchases))]
cov_2 = sum(cov_prep)/(len(sum_purchases)-1)

cov_2/(std_x_2*std_y_2)
```
